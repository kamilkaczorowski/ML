```python
import numpy as np
import  tensorflow as tf
```

```python
#name scope for variables
with tf.name_scope("variables"):
#creating matrixes with 2x3 and 3x2 dimensions
    matrix1 = tf.Variable(tf.random_normal((2, 3)))
    matrix2 = tf.Variable(tf.random_normal((3, 2)))

#name scope for calculations
with tf.name_scope("calculations"):
#storing matrixs multiplication in result tensor
    matrixResult = tf.matmul(matrix1, matrix2)

#running new session and processing calculations
with tf.Session() as sess:
#variables initialization
    sess.run(tf.global_variables_initializer())
    print("matrix1 dim: ", matrix1.shape[0], matrix1.shape[1])
    print(sess.run(matrix1))
    print("matrix2 dim: ", matrix2.shape[0], matrix2.shape[1])
    print(sess.run(matrix2))
    print("matrixResult dim: ", matrixResult.shape[0], matrixResult.shape[1])
    print(sess.run(matrixResult))
```

```python
#declaring two numpy arrays
a = np.random.rand(2,3)
b = np.random.rand(3,2)

#name scope for variables
with tf.name_scope("variables"):
#creating placeholders for x and y arrays
    x = tf.placeholder("float", [2, 3])            #no data here, it's just placeholder 
    y = tf.placeholder("float", [3, 2])            #no data here, it's just placeholder 

#name scope for calculations
with tf.name_scope("calculations"):
#creating tensor z for multiplication of x and y matrixes
    z = tf.matmul(x,y)

#running new session and processing calculations
with tf.Session() as session:
#feeding placeholder with numpy arrays and calculating result of multiplication
    result = session.run(z, feed_dict={x: a, y:b})
    print("x dim: ", x.shape[0], x.shape[1])
    print("y dim: ", y.shape[0], y.shape[1])
    print("z dim: ", z.shape[0], z.shape[1])
    print("z matrix:")
    print(result)
```

```python
# Creating x and y numpy arrays

x_np = np.random.rand(50, 1)
y_np = np.reshape(5 * x_np  + 2, (-1))
```

```python
# Number of nodes in hidden layers
hidden_nodes = 50

#name scope for variables
with tf.name_scope("inputs_variables"):
#creating placeholders for x and y arrayes
  x = tf.placeholder("float", [50,1])
  y = tf.placeholder("float", [50])

#name scope for hidden_layer
with tf.name_scope("hidden_layer"):
#creating variables for hidden layer
  W_h = tf.Variable(tf.random_normal([1,hidden_nodes]))    # 1 rows x n_hidden layers columns tensor with weight
    
#https://www.tensorflow.org/api_docs/python/tf/nn/relu
  x_h = tf.nn.relu(tf.matmul(x, W_h))                      # 50 rows x n_hidden layers columns tensor with output of hidden layers,
                                                           #calculated by using relu function

#name scope for output
with tf.name_scope("output"):
#creating variables for output
  W_out = tf.Variable(tf.random_normal([hidden_nodes, 1]))  # n_hidden layers rows x 1 columns tensor with weight
  y_pred = tf.transpose(tf.matmul(x_h, W_out))              # predicted values for y 50 rows

#name scope for error   
with tf.name_scope("loss"):
#calculating squered error
  lvec = (y - y_pred)**2
#calculating sum of squered error
  l = tf.reduce_sum(lvec)

#name scope for optimalization  
with tf.name_scope("optimalization"):
#Minimizing sum of squered error by using AdamOptimizer

#https://www.tensorflow.org/api_docs/python/tf/compat/v1/train/AdamOptimizer
  train_op = tf.train.AdamOptimizer(.001).minimize(l)

#Training model
with tf.Session() as sess:
  sess.run(tf.global_variables_initializer())
  for i in range(1000):
        loss, _ = sess.run([l, train_op], feed_dict={x: x_np, y: y_np})
        print("epoch: ", i+1, "sum of squered error: ", loss)
```
