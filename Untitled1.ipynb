```python
X_train_poly, X_test_poly, Y_train_poly, Y_test_poly = train_test_split(X, Y, test_size=0.3, random_state=43)
```

```python
reg_lin_poly = LinearRegression()
reg_lin_poly.fit(X_train_poly, Y_train_poly)
y_pred = reg_lin_poly.predict(X_test_poly)
mean_absolute_error(y_pred, Y_test_poly)
```

```python
hlp.plot_regression(X=X_train_poly, y= Y_train_poly, reg=reg_lin_poly)
```

```python
for d in [1,2,3,4,5,6,7,8]:
    reg_svm = SVR(kernel='poly', degree=d)
    reg_svm.fit(X_train, Y_train)
    hlp.plot_regression(X=X_train_poly, y= Y_train_poly, reg=reg_svm)
```

```python
reg_tree = DecisionTreeRegressor(criterion='mse')
reg_tree.fit(X_train, Y_train)
hlp.plot_regression(X=X_train_poly, y= Y_train_poly, reg=reg_tree)
```

```python
for d in range(1,8):
    reg_tree = DecisionTreeRegressor(criterion='mse', max_depth=d)
    reg_tree.fit(X_train, Y_train)
    hlp.plot_regression(X=X_train_poly, y= Y_train_poly, reg=reg_tree)
```

```python
for d in range(1,8):
    reg_tree = DecisionTreeRegressor(criterion='mse', min_samples_leaf=d)
    reg_tree.fit(X_train, Y_train)
    hlp.plot_regression(X=X_train_poly, y= Y_train_poly, reg=reg_tree)
```

```python
for e in [2,4,8,100,1000]:
    reg_rf = RandomForestRegressor(n_estimators = e, criterion='mse', min_samples_leaf=8)
    reg_rf.fit(X_train, Y_train)
    hlp.plot_regression(X=X_test, y= Y_test, reg=reg_rf)
```

```python
for e in [2,4,8,100,1000]:
    reg_rf = ExtraTreesRegressor(n_estimators = e, criterion='mse', min_samples_leaf=8)
    reg_rf.fit(X_train, Y_train)
    hlp.plot_regression(X=X_test, y= Y_test, reg=reg_rf)
```
